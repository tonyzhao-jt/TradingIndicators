```python
import matplotlib.pyplot as plt

# Data
categories = ['Successful\nConversion', 'Other']
values = [32, 988]  # 1020 total

# Colors
colors = ['#2E8B57', '#CCCCCC']

# Create pie chart
plt.figure(figsize=(7, 5))

# Plot
wedges, texts, autotexts = plt.pie(values, 
                                  colors=colors,
                                  autopct='%1.1f%%',
                                  startangle=90,
                                  textprops={'fontsize': 10})

# Style percentage text
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')

# Title
plt.title('Strategy Code Conversion Success Rate\n(Total: 1,020 strategies)', 
          fontsize=12, fontweight='bold', pad=15)

# Add legend on the right
plt.legend(wedges, ['Successful: 32 (3.1%)', 'Other: 988 (96.9%)'],
          loc='center left',
          bbox_to_anchor=(1, 0, 0.5, 1))

# Equal aspect ratio
plt.axis('equal')
plt.tight_layout()
plt.show()
```
